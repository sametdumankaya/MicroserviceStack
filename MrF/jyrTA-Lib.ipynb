```python
import numpy
import talib
import matplotlib.pyplot as plt
import numpy as np

close = numpy.random.random(100)
close2 = numpy.random.random(100)
plt.plot(close)
plt.plot(close2)
```

```python
output = talib.SMA(close)
plt.plot(output)
```

```python
from talib import MA_Type

upper, middle, lower = talib.BBANDS(close, matype=MA_Type.T3)
plt.plot(upper)
plt.plot(middle)
plt.plot(lower)
```

```python
output = talib.MOM(close, timeperiod=5)
plt.plot(output)
```
